```python
#Enter your input

s = '[({)]'
```

```python
mapping = {')': '(', '}': '{', ']': '['}  # Closing → Opening
stack = []        
```

```python
for char in s:
    if char in mapping:  # It's a CLOSING bracket
        top = stack.pop() if stack else '#'  # Get top (or dummy)
        if mapping[char] != top:  # Does it match?
            print('false')
    else:
        stack.append(char)  # It's OPENING → push
    
    if len(stack) == 0:  # All closed?
        print(True)
        
```
